```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
print(plt.style.available)
```

```python
plt.style.use('dark_background')
```

```python
fig, axes = plt.subplots(figsize=(8,8))
axes.plot(x,x+1,'r--')
axes.plot(x,x+2,'go-')
axes.plot(x,x+3,'c.-')
axes.plot(x,x+4,'b:')
```

## Pyplot

```python
fig,axes=plt.subplots(figsize=(8,8))
axes.plot(x,x+1,color='green',alpha=0.8,linewidth=5, linestyle='-',marker='o')
axes.plot(x,x+2,color='blue',linewidth=3, linestyle='dashed',marker='x')
axes.plot(x,x+3,color='#ff66',linewidth=10, linestyle='--',marker='v')
axes.plot(x,x+4,color='pink',linewidth=1, linestyle=':',marker='P')
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['INDIA', 'JAPAN', 'MEXICO', 'COLOMBIA', 'GERMANY']
population =[20000,20300,63040,59919,18538]
```

```python
plt.bar(country,population,width=0.5,color=['#E14D2A','#FD841F','#3E6D9C','#001253','#000620'])
plt.xticks(np.arange(5),('India','Japon','Mexico','Colombia','Alemania'),rotation=45)
plt.xlabel('Cantidad de gente')
plt.ylabel('Países')
plt.title('Cantidad de gente por país')
plt.tight_layout
plt.show()
```

```python
plt.barh(country,population)
plt.show()
```

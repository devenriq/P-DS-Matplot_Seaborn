```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(1,2,figsize=(10,5))
axes[0].plot(x,y,'b', label='$sin(x)$')
axes[0].set_title('Relación X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()

axes[1].plot(y,x,'g',label='$sin(y)$')
axes[1].set_title('Relacion Y - X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
axes[1].legend()

plt.tight_layout()

plt.show()
```

```python
plt.plot(x,y,label='sin(x)')
plt.title('Este es el título')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='lower left', bbox_to_anchor=(0.5,0.9))
```

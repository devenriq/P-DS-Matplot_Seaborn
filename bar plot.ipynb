```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country=['INDIA','JAPAN','MEXICO','COLOMBIA','GERMANY']
population = [10002001,1000230,3030020,10310200,1203102]
```

```python
plt.bar(country,population,width=0.4,color=['#E14D2A','#FD841F','#3E6D9C','#001253'])
plt.xticks(np.arange(5),('India', 'Japan', 'Mexico', 'Colombia', 'Germany'),rotation=45)
plt.title('Población por país')
plt.xlabel('Países')
plt.ylabel('Populacion')
plt.show()
```

```python
plt.barh(country,population)
plt.show()
```

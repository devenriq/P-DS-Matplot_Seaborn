```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
print(plt.style.available)
```

```python
plt.style.use('dark_background')
```

## Modificaciones Matlab

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1,'r--')
ax.plot(x,x+2,'b-')
ax.plot(x,x+3,'y.-')
ax.plot(x,x+4,'go:')
```

## Pyplot

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1,color='green')
ax.plot(x,x+2,color='red')
ax.plot(x,x+3,color='blue')
ax.plot(x,x+4,color='yellow')
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y,'b')
fig.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.6,0.9])
ax2 = fig.add_axes([0.15,0.55,0.3,0.4])

ax1.plot(x,y,'b')
ax2.plot(y,x,'r')
ax1.set_facecolor('yellow')
ax2.set_facecolor('gray')

fig2 = plt.figure()
ax1 = fig2.add_axes([0.1,0.1,0.6,0.9])
ax2 = fig2.add_axes([0.15,0.55,0.3,0.4])

ax1.plot(x,y,'g')
ax2.plot(y,x,'c')

fig.show()
fig2.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x,y,'b')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2)
ax1.plot(x,y,'b')
ax2.plot(y,x,'r')
plt.show()
```

```python
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(2,4)
ax1.plot(x,y,'g')
ax2.plot(x,np.cos(y),'b')
ax3.plot(y,np.cos(x),'r')
ax4.plot(x,np.tan(y),'g')

ax5.plot(x,y**2,'g')
ax6.plot(x,np.cos(y)**2,'b')
ax7.plot(y,np.cos(x)**2,'r')
ax8.plot(x,np.tan(y)**2,'g')

fig.tight_layout()
plt.show()
```
